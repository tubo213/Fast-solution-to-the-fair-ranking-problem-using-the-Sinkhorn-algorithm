```python
import japanize_matplotlib  # noqa: F401
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import torch

from src.optimizers._ot_utils import sinkhorn
```

## コスト行列の作成

```python
N = 5  # アイテム数、表示数
N_ITER = 10  # 反復回数
C = torch.rand(1, N, N) * torch.arange(1, 1 + N).view(1, N, 1)
# 0-1正規化
C = C / C.max()

# プロットの列数、行数
n_col = 5
n_row = N_ITER // n_col
if N_ITER % n_col != 0:
    n_row += 1
```

## $X = SA(X)$の更新を繰り返したときのXを可視化する

$X = SA(X)$という更新を繰り返したときのXを可視化する

$X$は費用行列のため、$X$と$SA(X)$の大小関係は逆になる傾向があるはず
そのため、$X = SA(X)$の更新を繰り返しても、$X$は振動して収束しないはず

```python
fig, ax = plt.subplots(n_row, n_col, figsize=(4 * n_col + 1, 4 * n_row))
ax: list[plt.Axes] = np.ravel(ax).tolist()
for iter in range(N_ITER):
    if iter == 0:
        sns.heatmap(
            C[0].numpy(),
            annot=False,
            fmt=".1f",
            cmap="Blues",
            ax=ax[0],
            vmin=0,
            vmax=1,
            cbar=False,
        )
        ax[0].set_title("コスト行列 C")
        pi = sinkhorn(C, 1, 1)
        title = "SA(C)"
    else:
        sns.heatmap(
            pi[0].numpy(),
            annot=False,
            fmt=".1f",
            cmap="Blues",
            ax=ax[iter],
            vmin=0,
            vmax=1,
            cbar=False,
        )
        ax[iter].set_title(title)
        ax[iter].set_xlabel("表示位置")
        ax[iter].set_ylabel("アイテム")
        pi = sinkhorn(pi, 1, 1)
        title = f"SA({title})"

fig.tight_layout()
# 図の右側にカラーバーを追加
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(ax[0].collections[0], cax=cbar_ax)
```

## $X$ = $SA(-X)$の更新を繰り返したときの$X$を可視化する

更新を繰り返したときの振動を避けるため、$X = SA(-X)$の更新を繰り返す

```python
fig, ax = plt.subplots(n_row, n_col, figsize=(4 * n_col + 1, 4 * n_row))
ax: list[plt.Axes] = np.ravel(ax).tolist()
for iter in range(N_ITER):
    if iter == 0:
        sns.heatmap(
            C[0].numpy(),
            annot=False,
            fmt=".1f",
            cmap="Blues",
            ax=ax[0],
            vmin=0,
            vmax=1,
            cbar=False,
        )
        ax[0].set_title("コスト行列 C")
        pi = sinkhorn(C, 1, 1)
        title = "SA(C)"
    else:
        sns.heatmap(
            pi[0].numpy(),
            annot=False,
            fmt=".1f",
            cmap="Blues",
            ax=ax[iter],
            vmin=0,
            vmax=1,
            cbar=False,
        )
        ax[iter].set_title(title)
        ax[iter].set_xlabel("表示位置")
        ax[iter].set_ylabel("アイテム")
        pi = sinkhorn(-pi, 1, 1)
        title = f"SA(-{title})"

fig.tight_layout()
# 図の右側にカラーバーを追加
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(ax[0].collections[0], cax=cbar_ax)
```

## 極端な $X$ に$X = SA(-\log(X))$ を適用したときの $X$ を可視化する

0列目が $1$ でそれ以外が微小な値であるような $X$ に $X = SA(-\log(X))$ を適用したときの $X$ を可視化する

```python
C = torch.zeros_like(C)
C[:, :, 0] = 1
C[:, :, 1:] = 1e-5
```

```python
N_ITER = 2
n_col = 2
n_row = N_ITER // n_col
if N_ITER % n_col != 0:
    n_row += 1

fig, ax = plt.subplots(n_row, n_col, figsize=(4 * n_col + 1, 4 * n_row))
ax: list[plt.Axes] = np.ravel(ax).tolist()
for iter in range(N_ITER):
    if iter == 0:
        sns.heatmap(
            C[0].numpy(),
            annot=True,
            fmt=".1f",
            cmap="Blues",
            ax=ax[0],
            vmin=0,
            vmax=1,
            cbar=False,
        )
        ax[0].set_title("初期値")
        ax[0].set_xlabel("表示位置")
        ax[0].set_ylabel("アイテム")
        pi = sinkhorn(C, 1, 1, apply_kernel=False)
        title = "SA(初期値)"
    else:
        sns.heatmap(
            pi[0].numpy(),
            annot=True,
            fmt=".1f",
            cmap="Blues",
            ax=ax[iter],
            vmin=0,
            vmax=1,
            cbar=False,
        )
        ax[iter].set_title(title)
        ax[iter].set_xlabel("表示位置")
        ax[iter].set_ylabel("アイテム")
        pi = sinkhorn(pi, 1, 1, apply_kernel=False, eps=1)
        title = f"SA({title})"

fig.tight_layout()
```
